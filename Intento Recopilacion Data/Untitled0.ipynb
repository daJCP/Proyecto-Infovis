```python
# Rendimiento general equipo
season = '2022-2023'
league = 'Premier-League'
url = f'https://fbref.com/en/comps/9/{season}/{season}-{league}-Stats'
headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:78.0) Gecko/20100101 Firefox/78.0'}
response = requests.get(url, headers=headers)


soup = BeautifulSoup(response.text, "html.parser")
q = soup.find_all('tr')

df_season_stats = 0
for i in range(0, 21):
  elements = q[i].find_all(['td', 'th'])
  if i == 0:
    texts_tabla = [element.get_text() for element in elements]
    texts_tabla.append('link')
    df_season_stats = pd.DataFrame(columns=texts_tabla)
  else:
    texts_tabla = [element.get_text() for element in elements]
    try:
      texts_tabla.append(elements[1].find_all('a')[0]['href'])
    except:
      texts_tabla.append(None)

    df_season_stats.loc[len(df_season_stats)] = texts_tabla
```
